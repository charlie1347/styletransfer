```python
def show_rsp_image(resp):
    return Image.open(BytesIO(resp.content)) 
```

```python
# Get styles
url = "http://styletransfer.azurewebsites.net/api/styles"
styles = requests.get(url).json()
```

```python
for style in styles.keys():
    url = "http://styletransfer.azurewebsites.net/api/uploader/{0}".format(style)
    files = {'imagefile': open('image.jpg', 'rb')}
    %time out = requests.post(url, files=files)
    style
    show_rsp_image(out)
```
